```python
# Positional Argument
```

```python
def show(name,age):
  print(name,age)

show("aditi", 18)
```

```python
def show(name,age):
  print(f"I am {name} , and I am {age} years old")

show("Aditi",18)
```

```python
def show(age,name):
  print(f"I am {age}years old and my name is {name}")

show(18,"Aditi")
```

```python
def show(name,age):
  print(f"I am {name} , and I am {age} years old")

show(18, "Aditi")
```

```python
# Keyword argument
```

```python
def show(name,age):
  print(f"I am {name} , and I am {age} years old")

show(age=18, name= "Aditi")
```

```python
def show(name,age,marks):
  print(name,age,marks)

show(marks=90, name="Aditi", age=18)
```

```python
# **Kwargs (key- value)
```

```python
def info(**data):
  print(data)

info(name="Aditi",age=18)
```

```python
def info(**data):
  print(data)

info(name="Aditi",age=18, subject="python", marks=98, branch="Information Technology")
```

```python
def add(*nums):
   print(sum(nums))


add(10,20,30.45)

```

```python
def mul(*nums):
  result = 1
  for num in nums:
    result *= num
  print(result)

mul(10,20,30)
```
