```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4]
y= [10,20,25,30]
```

```python
#Basic line plot
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)

plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
```

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1,2,3]
y1=[2,4,6]
y2=[1,3,5]

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x =['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
y = [10,20 ,30,50, 60, 70, 90]
```

```python
plt.plot(x,y)
plt.title("Days vs Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
categories = ["A", "B", "C"]
values= [10,20,15]

plt.bar(categories, values)

plt.xlabel("Categories")
plt.ylabel("values")

plt.title("Simple Bar Chart")
```

```python
plt.barh(categories, values)
plt.show()
```

```python
labels = ["A", "B", "C"]
sizes = [30,40,30]
```

```python
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
plt.title("Simple Pie chart")
plt.show()
```

```python
plt.pie(sizes, labels = labels, autopct='%.2f')
plt.title("Simple Pie chart")
plt.show()
```

```python
plt.pie(sizes, labels = labels, autopct='%1.0f%%')
plt.title("Simple Pie chart")
plt.show()
```

```python
scores = [ 50,55,60,66,70,77,88,89,90,99]

counts, bins, patches = plt.hist(scores, bins=5, edgecolor='black',color='skyblue')

plt.bar_label(patches)

plt.title("Histogram of Student Scores")

plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```
