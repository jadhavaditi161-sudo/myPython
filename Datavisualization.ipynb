```python
sns.kdeplot(data=data, x="sepal_length", fill=False)
```

```python

plt.title("univariate distribution")
plt.show()
```

```python
sns.kdeplot(data=data, x="sepal_length", fill=True)
plt.title("univariate distribution")
plt.show()
```

```python
sns.kdeplot(
    data=data,
    x="sepal_length",
    y="sepal_width",
    fill=True
)
plt.title("biavariate distribution")
plt.show()
```

```python
#jointplot
```

```python
sns.jointplot(
    data=data,
    x="sepal_length",
    y="sepal_width",
    kind="scatter"
)
```

```python
#pairplot
```

```python
sns.pairplot(data)
plt.show()
```
