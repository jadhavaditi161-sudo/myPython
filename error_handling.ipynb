```python
try:
  x= 10/0
except ZeroDivisionError:
  print("Cannot divide by zero")
```

```python
# Multiple except blocks
# different errors can be handled in different ways using multiple except blocks.
```

```python
try:
  num = int("abc")
  result = 10/ num
except ValueError:
  print("Conversion Failed")
except ZeroDivisionsError:
  print("Division by zero error")
```

```python
try:
  x = 10/0
  result = 10/x
except ValueError:
  print("Conversion Failed")
except ZeroDivisionError:
  print("Division by zero error")
```

```python
# multiple exceptions in one except
# You can handle more than one error type in a single block.
```

```python
try:
  x = int("abc")
except (ValueError, TypeError):
  print("Value or Type error occurred")
```

```python
try:
  x=10/0
  result = 10/ x
except ValueError:
  print("Conversion Failed")
except ZeroDivisionError:
  print("Division by zero error")
```

```python
# Generic Exception
# Catches any kind of error when you don't specify the exact type.
```

```python
try:
  x = int("abc")
except Exception as e:
  print("Error occurred:",e)
```

```python
# else block
# Runs only when no error occurs in the try block
```

```python
try:
  x = 10/2
except ZeroDivisionEror:
  print("Error")
else:
  print("Success:",x)
```

```python
# Finally block
# Runs every time, whether an error happens or not,often used for cleanup
```

```python
try:
  x = 10/ 0
except ZeroDivisionError:
  print("Error Occured")
finally:
  print("this always run")
```

```python
# raise Keyword
# Used to manually create and throw an error.
```

```python
def check_age(age):
  if age < 18:
    raise ValueError("Age must be 18+")
  return "Allowed"

try:
  check_age(15)
except ValueError as e:
  print(e)
```

```python
# Custom Exception
# You can create your own error type for specific situations.
```

```python
class CustomError(Exception):
  pass
def test():
  raise CustomError("This is a  custom error")
try:
  test()
except CustomError as e:
  print(e)
```

```python
# assert statement
# Checks a condition; if false, it raises an error.
```

```python
x = -5
try:
  assert x > 0, "X must be positive"
except AssertionError as e:
  print(e)
```

```python
# Nested Try- except
# A try-except block placed inside another.
```

```python
try:
  try:
    x = int("abc")
  except ValueError:
    print("Inner error")
    raise
except Exception:
  print("Outer handler caught it")
```

```python
# File Handling with exceptions
# Used to handle errors while working with files.
```

```python
try:
  f= open("file.txt", "r")
  data = f.read()
except FileNotFoundError:
  print("File not Found")
finally:
    try:
      f.close()
    except:
       pass
```
