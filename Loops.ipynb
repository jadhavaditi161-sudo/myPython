```python
# A while loop in Python runs the same block of code over and over again as long  as a specific conditions is True.
```

```python
# break
```

```python
# The break statement in python instantly stops and exits a loop,even if the loop's condition is still True
```

```python
# Stop loop at 5
```

```python
for i in range(1,11):

  if i == 5:
    break

  print(i)
```

```python
# ATM exit program
```

```python
while True:
  amount = int(input("Enter Amount (0 to Exit):"))

  if amount == 0:
    break

  print("Amount Entered:", amount)
```

```python
# continue
```

```python
# The continue statement in Python skips the rest of the code inside the current loop iteration and jumps straight to the next turn
```

```python
# Skip number 5
```

```python
for i in range(1,11):

  if i == 5:
    continue

  print(i)
```

```python
# print only odd numbers
```

```python
for i in range(1,21):

  if i % 2 == 0:
    continue

  print(i)
```
